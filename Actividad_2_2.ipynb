### **FACTURACIÓN**

En este caso uitilizamos sustituimos los valores nulos por un string ya establecido "--", con el fin de identificar cuando no contamos con dicho valor, a pesar de que la clave del vendedor es del tipo string, no consideramos relevante que se mantenga así, ya que no se utilizara para realizar operaciones, sino solo para conocer que vendedor realizo la operación y en la fecha de cancelación al ser tantos datos nulos y al observar que la fecha de entrega no es del tipo datetime, no consideramos que relevante mantener el tipo datetime, por lo cual decidimos usar ese string, ya que el poner otra fecha nos diría que si se cancelo la factura, lo cual no sucedio.

```python
fact.info()
```

```python
fact['CVE_VEND'] = fact['CVE_VEND'].fillna('--')
fact['FECHA_CANCELA'] = fact['FECHA_CANCELA'].fillna('--')
fact['FECHA_ENT'] = fact['FECHA_ENT'].fillna('--')
```

```python
fact.info()
```

### **Detalles de precios 2022**

En este caso de igual manera utilice el string "--" el cual solo servira para que los datos nulos no afecten el analisis, ya que este string nos dará a conocer si el nombre del vendedor falta.

```python
det.info()
```

```python
det['NOMBRE'] = det['NOMBRE'].fillna('--')
```

```python
det.info()
```

### **Gastos y costos 2020**

En este caso todas las columnas con datos faltantes que tuvieran tipos de datos object, sus datos nulos fueron sustituídos por el string "--", de igual manera solo para poder identificar que es un dato faltante y que no afecte el análisis.
Por otra parte, en el tipo de cambio, se sustituyeron los datos nulos por el valor 1, ya que esto significa que se maneja la misma moneda, es decir en pesos.
Finalmente, para el importe y el iva se utilizó la mediana, con el fin de no perder información relevante y que los valores atipicos no afectaran el analisis como en el caso del promedio.

```python
gyc20.info()
```

```python
gyc20['FOLIO'] = gyc20['FOLIO'].fillna('--')
gyc20['GASTO'] = gyc20['GASTO'].fillna('--')
gyc20['TC'] = gyc20['TC'].fillna(1)
gyc20['IMPORTE'] = gyc20['IMPORTE'].fillna(round(gyc20["IMPORTE"].median(),1))
gyc20['IVA'] = gyc20['IVA'].fillna(round(gyc20["IVA"].median(),1))
gyc20['TIPO'] = gyc20['TIPO'].fillna('--')
gyc20['POLIZA'] = gyc20['POLIZA'].fillna('--')
```

```python
gyc20.info()
```

### **Gastos y costos 2021**

En este caso las tres columnas en las que tenemos valores faltantes son del tipo de dato objecto, por lo que solo queremos identificarlos y que no nos afecten el análisis, por lo que se sustituyen con el string "--"

```python
gyc21.info()
```

```python
gyc21['FOLIO'] = gyc21['FOLIO'].fillna('--')
gyc21['MP'] = gyc21['MP'].fillna('--')
gyc21['POLIZA'] = gyc21['POLIZA'].fillna('--')
```

```python
gyc21.info()
```

### **Gastos y costos 2022**

Para este dataset tenemos 3 columnas con tipo de dato object, las cuales son Folio, MP y Poliza, de modo que como en los casos anteriores se les sustituye con "--", mientras que el tipo de cambio se sustituye a los valores nulos con 1 y el el caso de la columna otros al no tener un solo dato no nulo, se borra.

```python
gyc22.info()
```

```python
gyc22['Folio'] = gyc22['Folio'].fillna('--')
gyc22['MP'] = gyc22['MP'].fillna('--')
gyc22['TC'] = gyc22['TC'].fillna(1)
gyc22['Poliza'] = gyc22['Poliza'].fillna('--')
gyc22 = gyc22.dropna(axis = 1)
```

```python
gyc22.info()
```

### **Gastos y costos 2023**

De igual manera que en el dataset anterior las columnas con tipo de dato object fueron sustituidas con "--", a excepción de la columna FP que tenía objeto flotante, la cual recibió la misma forma de sustitución de datos nulos, esto ya que, FP es un codigo que recibe cada municipio que es MP, por lo que no tiene gran relevancia para realizar operaciones o un análisis cuantitativo, por lo que no es relevante mantener el tipo de dato, así mismo, la columna otros se elimino.

```python
gyc23.info()
```

```python
gyc23['Folio'] = gyc23['Folio'].fillna('--')
gyc23['TIPO GASTO'] = gyc23['TIPO GASTO'].fillna('--')
gyc23['MP'] = gyc23['MP'].fillna('--')
gyc23['FP'] = gyc23['FP'].fillna('--')
gyc23['Poliza'] = gyc23['Poliza'].fillna('--')
gyc23 = gyc23.dropna(axis = 1)
```

```python
gyc23.info()
```
