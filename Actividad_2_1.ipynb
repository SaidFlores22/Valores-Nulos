### **Salon ventas**

Para este caso utilice el promedio ya que no hay tanta diferencia entre promedio y el percentil 50%, es decir la mediana, por lo que los valores atipicos no influyen tanto, de esto modo no perdere datos en caso de querer saber que canal de venta es el principal de la empresa.

```python
data["salon_ventas"] = data["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
data
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

### **Tarjeta de debito, trajeta de crdito y otros medios**

En estos tres caso, decid utilizar el forward fill, ya que al ser ventas por medios de pago, es probable que las ventas de un da antes sean muy similares a las del da anterior debido a la temporada, en especial hablando del medio de pago, por lo tanto, decid implementar esa sustitucin, con el fin de que el anlisis no se vea afectado por valores atipicos.

```python
data["tarjetas_debito"] = data["tarjetas_debito"].fillna(method = "ffill")
data["tarjetas_credito"] = data["tarjetas_credito"].fillna(method = "ffill")
data["otros_medios"] = data["otros_medios"].fillna(method = "ffill")
data
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

### **Subtotal ventas de alimentos y bebidas**

En este caso al haber muchos datos nulos, considerando la cantidad de datos que tenemos, no considere viable utilizar la mediana o el promedio, ya que se podran ver sesgados, debido a algn valor atipico, al haber menos valores no nulos, por lo tanto decid utilizar el metodo backward ya que podra haber similitud en las ventas del da siguiente, aunque anlizando bien el dataset, mi recomendacin sera realizar la suma de las ventas de todos los alimentos y bebidas de la fila en la que se encuentra el dato nulo, pero con el fin de utilizar los metodos vistos, tome el ya mencionado.

```python
data["subtotal_ventas_alimentos_bebidas"] = data["subtotal_ventas_alimentos_bebidas"].fillna(method = "bfill")
data
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

### **Bebidas, almacen, panaderia, lacteos, carnes, verduleria, alimentos preparados, calzado, electronicos y otros**

En estos casos al anlizar sus promedios, sus medianas, as como sus valores mximos y mnimos, puedo identificar que en la mayora si hay gran diferencia entre los valores maximos y el percentil 75, es decir que si llega a haber valores atipicos, asems que el promedio si es considerablemente mayor a la mediana, por lo tanto, considero viable utilizar la mediana para sustituir los valores nulos, esto debido a que adems son muy pocos valores faltantes por columna, incluso en algunos casos es solo un valores nulo, por lo que no veo necesario utilizar backward o forward fill, adems, no creo que sea recomendable sustituir por algn valor establecido como 0, ya que se convertira en un valor atipico que podra sesgar el anlisis, as mismo no me gustara eliminar las filas, ya que se perdera informacin valiosa, que de por si es poca.

```python
data["bebidas"] = data["bebidas"].fillna(round(data["bebidas"].median(),1))
data["almacen"] = data["almacen"].fillna(round(data["almacen"].median(),1))
data["panaderia"] = data["panaderia"].fillna(round(data["panaderia"].median(),1))
data["lacteos"] = data["lacteos"].fillna(round(data["lacteos"].median(),1))
data["carnes"] = data["carnes"].fillna(round(data["carnes"].median(),1))
data["verduleria_fruteria"] = data["verduleria_fruteria"].fillna(round(data["verduleria_fruteria"].median(),1))
data["alimentos_preparados_rotiseria"] = data["alimentos_preparados_rotiseria"].fillna(round(data["alimentos_preparados_rotiseria"].median(),1))
data["indumentaria_calzado_textiles_hogar"] = data["indumentaria_calzado_textiles_hogar"].fillna(round(data["indumentaria_calzado_textiles_hogar"].median(),1))
data["electronicos_articulos_hogar"] = data["electronicos_articulos_hogar"].fillna(round(data["electronicos_articulos_hogar"].median(),1))
data["otros"] = data["otros"].fillna(round(data["otros"].median(),1))

data
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```
